```python
# Defining the function Graphs for the Working of the 'Cars' Dataset  which we choose for the Exploratory Data Analysis.

cars=pd.read_csv('cars.csv')
def Graphs(data,cols='default',directory=os.getcwd()): # Three Parameters inside the Function.
     # It saves in the Directory 
    # Appending the list of all column names of the cars dataset in the cols list.
    col = [] 
    if cols=='default':
        for i in data:
            col.append(i)
    else:
        col=cols
    # Running a function inside the columns list created. 
    for i in (col):
        # The visualitsation  of the 'objects' in the cars dataset apart from other data types. 
        if data.dtypes[i]=='object':
             # The Bar graphs best Represents the EDA of the categorical Datatype.  
            data[i].value_counts().plot(kind = "bar",figsize = (5,5),color ='coral',fontsize =12)
            plt.xlabel(i,fontsize=18) # Adding the X label for the visual.
            plt.savefig(directory+'\\'+i+'.png', dpi=1000, bbox_inches='tight') # The Images created are Easy are saved in the PNG format in the file.  
            plt.show()  
            # The visualisation  of the Datatypes other than the  object data type in the cars dataset is done in the else part.     

        else:
            # The Subplots are made in the cars dataset apart from the objects for the Data Visualization.

            # Two plots are made one is the Box - Plot and other is the Histogram for the subplots  
            plt.subplot(2,1,1)
            plt.boxplot(data[i],vert = False)# Box - plot of int and float object type            
            plt.title("Boxplot of"+i)
            plt.xlabel(i) # Presenting the X label.
            plt.subplot(2, 1, 2)
            
            plt.hist(data[i],bins=6)
            plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.57)
            plt.title("Histogram of "+i) # Adding the histogram in the Subplots
            plt.xlabel(i)
            plt.ylabel("count") 
            plt.savefig(directory+'\\'+i+'.png', dpi=100,bbox_inches='tight') # Saving the  png file in the  directory is made
            plt.show()
```

```python
Graphs(cars)
```
